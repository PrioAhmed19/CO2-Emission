```python
specific_string = 'Afghanistan'
afg = df[df['Country Name'] == specific_string]
afg
```

```python
columns_to_exclude = ['Country Code', 'Country Name','Indicator']
afg_1 = afg.drop(columns=columns_to_exclude)
```

```python
afg_1
afg_1.drop([5, 6,7], inplace=True)
```

```python
afg_train1=afg_1.T
```

```python
afg_train1
```

```python
afg_train2 = afg_train1.replace(r'\.\.', np.nan, regex=True)
null_counts = afg_train2.isnull().sum()
null_counts
```

```python
for column in afg_train2.columns:
    if null_counts[column] > 0:
        afg_train2[column] = afg_train2[column].fillna(afg_train2[column].rolling(window=len(afg_train2), min_periods=1, center=True).mean())
```

```python
null_counts = afg_train2.isnull().sum()
null_counts
```
